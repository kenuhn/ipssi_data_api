```python
import pandas as pd
import plotly.express as px
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
playList = pd.read_csv("playlist_2010to2022.csv")
```

```python
top_des_genre_par_an = playList[(playList['track_popularity']>88) & (playList['year']>=2010) & (playList['year']<=2022) ].groupby(['year','artist_genres']).sum(numeric_only=True)[['track_popularity']]
top_des_genre_par_an = top_des_genre_par_an.reset_index()
top_des_genre_par_an = top_des_genres_par_an.replace([])
```

```python
#plt.figure(figsize=(6, 480))
plt.title("top des Genres par Année (2010-2022)")
sns.barplot(top_des_genre_par_an,x = 'year',y='track_popularity',hue='artist_genres',width = 5)
legend = plt.legend(title='artist_genres', loc='center', bbox_to_anchor=(1.7,0.5),fontsize=8)
legend_title = legend.get_title()
legend_title.set_fontsize('x-large')
plt.savefig('C:/Users/utilisateur/Desktop/cours_ipssi/project/top_des_genres_2010_2022.png')
```

```python
plus_dn_par_gn = playList.groupby('artist_genres').sum(numeric_only = True)[['danceability']]
plus_dn_par_gn = plus_dn_par_gn.sort_values(by='danceability', ascending = False).head(10).reset_index()
plus_dn_par_gn
```

```python
plt.figure(figsize=(15,10))
plt.title("les_genres_plus dansantes")
#sns.barplot(x="artist_genres", y="danceability", data=plus_dn_par_gn, estimator=sum, palette="husl",legend=False,hue="artist_genres")
col = sns.color_palette("husl",10)
plt.bar(range(len(plus_dn_par_gn['artist_genres'])),plus_dn_par_gn['danceability'],align="center",color=col)
plt.xticks(range(len(plus_dn_par_gn['artist_genres'])),plus_dn_par_gn['artist_genres'],rotation=90)
```
